             Stack implementation

```python

class Stack:
    
    def __init__(self):
        self.stack=[]
    
    def push(self,item):
        '''
          Description: 
                this function is pushing element into stack
          Parameters: 
                element(obj): element 
          Return : 
                None
        '''
        self.stack.append(item)
        print(f"Pushed {item} in to stack")

    def size(self):
        '''
          Description: 
                this function is returning size of stack
          Parameters: 
                None
          Return : 
                size(int): returns size of stack
        '''
        return len(self.stack)
    
    def isempty(self):
        '''
          Description: 
                this function is returns boolean values if it empty return true or false
          Parameters: 
                None
          Return : 
                Returns boolean values
        '''
        return len(self.stack)==0
   
    def peek(self):
        '''
          Description: 
                this function is returns the top element in stack
          Parameters: 
                None
          Return : 
                Returns the top element in stack
        '''
        if not self.isempty():
            return self.stack[-1]
        
        else:
            print("Stack is empty")
            return None
    
    def pop(self):
        '''
          Description: 
                this function is removing top element in stack
          Parameters: 
                None
          Return : 
                Returns the top element in stack 
        '''
        if not self.isempty():
            item=self.stack.pop()
            print(f"Popped {item} from stack")
        else:
            print("stack dont have any element to pop")
            return None
    
    def display(self):
        '''
          Description: 
                this function is displaying element in stack
          Parameters: 
                None
          Return : 
                None
        '''
        if self.isempty():
            print("stack dont have any element to display")

        else:
            print("current stack :",self.stack)


def main():
    stack_1=Stack()
    stack_1.push(10)
    stack_1.push(20)
    stack_1.push(30)
    stack_1.push(40)
    
    print(stack_1.size())
    print(stack_1.isempty())
    print(stack_1.peek())
    print(stack_1.display())
    print(stack_1.pop())
    print(stack_1.display())


if __name__=="__main__":
    main()

```

2.Reverse a string using stack

```python

class Stack:
    
    def __init__(self):
        self.stack=[]
    
    def push(self,item):
        '''
          Description: 
                this function is pushing element into stack
          Parameters: 
                element(obj): element 
          Return : 
                None
        '''
        self.stack.append(item)
    
    def isempty(self):
        '''
          Description: 
                this function is returns boolean values if it empty return true or false
          Parameters: 
                None
          Return : 
                Returns boolean values
        '''
        return len(self.stack)==0
    
    def pop(self):
        '''
          Description: 
                this function is removing top element in stack
          Parameters: 
                None
          Return : 
                Returns the top element in stack 
        '''
        if not self.isempty():
            item=self.stack.pop()
            return item
        else:
            print("stack dont have any element to pop")
            return None
        
        def __str__(self):
            return str(self.stack)
     

def reverse_string_using_stack(string_original):
    '''
          Description: 
                this function is reversing string 
          Parameters: 
                None
          Return : 
                Returns the after reverse string
        '''
    stack=Stack()

    for char in string_original:
        stack.push(char)

    reverse_string=''

    while not stack.isempty():
        reverse_string=reverse_string+stack.pop()

    return reverse_string
    

def main():

    string_original="Venkatesh"
    print("Original string:",string_original)
    print("After reversing string:",reverse_string_using_stack(string_original))


if __name__=="__main__":
    main()
```
